```python
import pandas as pd
```

```python
node_num = 9
epochs = 2000
batch_size=32
```

```python
baseline = pd.read_csv('./dataset/titanic3.csv')
```

```python
baseline.info()
```

```python
baseline.describe()
```

```python
from keras import Sequential
from keras.layers import Dense
from keras.initializers import RandomUniform

NN = Sequential()
NN.add(Dense(units = node_num, kernel_initializer = RandomUniform(), activation = 'sigmoid', input_dim = 5))
NN.add(Dense(units = 1, kernel_initializer = RandomUniform(), activation = 'sigmoid'))
NN.summary()
```

```python
baseline
```

```python
feature = ['pclass', 'age', 'fare', 'sibsp', 'parch']
X = baseline[feature].copy()
Y = baseline['survived']
print(X, Y)
```

```python
mean_age = X['age'].mean(skipna=True)
mean_fare = X['fare'].mean(skipna=True)
X['fare'] = X['fare'].fillna(mean_fare)
X['age'] = X['age'].fillna(mean_age)
```

```python
X = X.dropna()
Y = Y.dropna()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
NN.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])
NN.fit(x_train, y_train, batch_size=batch_size, epochs=epochs)
```

```python
NN.save(f'baseline_{node_num}_{epochs}_{batch_size}.keras')  
```

```python
#from keras.models import load_model
#NN= load_model('baseline.keras')
```

```python
from sklearn.metrics import accuracy_score, recall_score, f1_score, precision_score, classification_report
y_pred = NN.predict(x_test)
y_pred = [1 if item > 0.5 else 0 for item in y_pred]
print(y_pred)

accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
precision = precision_score(y_test, y_pred, average='weighted')

```

```python
file_name = f'model_evaluation_baseline.txt'
with open(file_name, 'a') as f:
    f.write('**************************')
    f.write(f'node num: {node_num}\n')
    f.write(f'epochs: {epochs}\n')
    f.write(f'batch_size: {batch_size}\n')
    f.write(f'accs: {accuracy:.4f}\n')
    f.write(f'recall: {recall:.4f}\n')
    f.write(f'F1 score: {f1:.4f}\n')
    f.write(f'precision: {precision:.4f}\n')
    
    report = classification_report(y_test, y_pred)
    f.write("\nreport:\n")
    f.write(report)

print(f"Evaluation results saved to {file_name}")
```
