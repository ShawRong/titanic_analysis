```python
import pandas as pd
```

```python
node_num = 9 
batch_size = 32
epochs = 1000
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('./extend.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
df
```

```python
feature = ['Class', 'Sex', 'Real_Fare', 'Family', 'Title', 'Predict_Cabin', 'General Region', 'Embarked', 'Cabin Position']
X = df[feature].copy()
Y = df['Survived']
print(X, Y)
```

```python
X.isna().sum()
```

```python
mask = X.isna().any(axis=1)
X = X[~mask]
Y = Y[~mask] 
```

```python
X.isna().sum()
```

```python
Y.isna().sum()
```

```python
import numpy as np
sex = X['Sex']
X['Sex'] = np.where(sex == 'Male', 1, 0)
```

```python
X['Embarked'].unique()
```

```python
em = X['Embarked']
X['Embarked'] = np.where(em == 'C', 0, np.where(em == 'Q', 1, 2))
```

```python
X['Predict_Cabin'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder


label_encoder = LabelEncoder()
X['Title'] = label_encoder.fit_transform(X['Title'])
X['General Region'] = label_encoder.fit_transform(X['General Region'])
X['Predict_Cabin'] = label_encoder.fit_transform(X['Predict_Cabin'])
```

```python
print(X['Title'].unique())
print(X['General Region'].unique())
print(X['Predict_Cabin'].unique())
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
from keras import Sequential
from keras.layers import Dense
from keras.initializers import RandomUniform

NN = Sequential()
NN.add(Dense(units = node_num, kernel_initializer = RandomUniform(), activation = 'sigmoid', input_dim = 9))
NN.add(Dense(units = 1, kernel_initializer = RandomUniform(), activation = 'sigmoid'))
NN.summary()
```

```python
NN.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])
NN.fit(x_train, y_train, batch_size=batch_size, epochs=epochs)
```

```python
import time
from datetime import datetime

current_timestamp = time.time()
dt_object = datetime.fromtimestamp(current_timestamp)
formatted_time = dt_object.strftime('%Y%m%d%H%M%S')
```

```python
NN.save(f'plus_{node_num}_{epochs}_{batch_size}.keras')  
```

```python
#from keras.models import load_model
#NN= load_model(f'refine_{node_num}.keras')
```

```python
from sklearn.metrics import accuracy_score, recall_score, f1_score, precision_score, classification_report
y_pred = NN.predict(x_test)
y_pred = [1 if item > 0.5 else 0 for item in y_pred]
print(y_pred)

accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
precision = precision_score(y_test, y_pred, average='weighted')
```

```python
file_name = f'model_evaluation.txt'
with open(file_name, 'a') as f:
    f.write('**************************')
    f.write(f'node num: {node_num}\n')
    f.write(f'epochs: {epochs}\n')
    f.write(f'batch_size: {batch_size}\n')
    f.write(f'accs: {accuracy:.4f}\n')
    f.write(f'recall: {recall:.4f}\n')
    f.write(f'F1 score: {f1:.4f}\n')
    f.write(f'precision: {precision:.4f}\n')
    
    report = classification_report(y_test, y_pred)
    f.write("\nreport:\n")
    f.write(report)

print(f"Evaluation results saved to {file_name}")
```
